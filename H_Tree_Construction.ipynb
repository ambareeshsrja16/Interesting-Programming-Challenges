# H-Tree Construction

An H-tree is a geometric shape that consists of a repeating pattern resembles the letter “H”.

It can be constructed by starting with a line segment of arbitrary length, drawing two segments of the same length at right angles to the first through its endpoints, and continuing in the same vein, reducing (dividing) the length of the line segments drawn at each stage by √10.

Here are some examples of H-trees at different levels of depth:

<img src = 'h_tree.PNG'>


Write a function draw_Tree that constructs an H-tree, given its center (x and y coordinates), a starting length, and depth. Assume that the starting line is parallel to the X-axis.

Use the function drawLine provided to implement your algorithm. In a production code, a drawLine function would render a real line between two points. However, this is not a real production environment, so to make things easier, implement drawLine such that it simply prints its arguments (the print format is left to your discretion).

Analyze the time and space complexity of your algorithm. In your analysis, assume that drawLine's time and space complexities are constant, i.e. O(1).

## Solution 1

```python
import itertools as it
import matplotlib.pyplot as plt
%matplotlib inline

def draw_Tree(length, centre_x, centre_y, depth=1):

    fig = plt.figure()

    def drawLine(x1, y1, x2, y2):
        """given (x1, y1), (x2, y2) draws a line between them"""
        x_s, y_s = [x1, x2], [y1, y2]    
        plt.plot(x_s, y_s, marker = 'o') 

    def draw_H(length, centre_x, centre_y):
        """draw one H at centre_x, centre_y"""
        l1_end_1 = (centre_x - length/2, centre_y)
        l1_end_2 = (centre_x + length/2, centre_y)

        l2_end_1 = (centre_x - length/2, centre_y + length/2)
        l2_end_2 = (centre_x - length/2, centre_y - length/2)

        l3_end_1 = (centre_x + length/2, centre_y + length/2)
        l3_end_2 = (centre_x + length/2, centre_y - length/2)

        drawLine(*l1_end_1 , *l1_end_2)
        drawLine(*l2_end_1 , *l2_end_2)
        drawLine(*l3_end_1 , *l3_end_2)

        return l2_end_1, l2_end_2, l3_end_1, l3_end_2

    #First H
    corners = list(draw_H(length, centre_x, centre_y)) # [ (0,0), (0,0), (0,0), (0,0),]
    depth -=1
    length = length / (10**0.5)
    
    #Subsequent H's
    while depth>0:
        temp_corners = []
        for corner in corners: #corner is a tuple (x,y)
            temp_corners.extend(list(draw_H(length, corner[0], corner[1])))
        corners = temp_corners
        depth -=1
        length = length / (10** 0.5)
```

```python
draw_Tree(100, 0, 0, depth=1)
```

```python
draw_Tree(100, 0, 0, depth=2)
```
